### Background

Our GoPro data collection consisted of 1080p video recording with single 1440p images taken roughly five seconds apart. Because the photos are a higher resolution than the video, it would be prefereable to use the photos for specimen measurement. However, it is difficult to determine the altitude that the photos were taken using our vehicle flight logs. In order to more easily determine the elevation from which particular photos were taken, we created a set of Python functions to match each photo with its closest-analogue video frame. This decreased the effort needed to align our flight logs (which contain altitude informaiton) to a particular image. Once the lineup is complete, the altitude at which the photo was taken can be determined using the video file as an intermediary between the image and the vehicle logs.

The <b>csvsweep()</b> function takes an input of a .csv file outlining approximate frame locations of photos. These were calculated based on the assumed locations of photos if they were taken exactly five seconds apart (as described by the camera manufacturer). The csv being input into the function should match the format of the example csv included in this repository

For our video evaluations in the <b>findframe()</b> function, we stripped our GoPro video of audio stream using ffmpeg. This was due to an issue with OpenCV that caused early video frames to be misread during a sweep.

```python
# Import all packages needed to analyze data - run this before anything else
import scipy, numpy, cv2
import matplotlib as mpl
import matplotlib.style as style
import csv
import glob
from matplotlib import pyplot as plt
```

```python
# run this cell before cells below it so jupyter can identify csvsweep() is when called

def csvsweep(path = '/home/devynn/github/Github/omurasWhaleAnalysis/example_data/csv/lineup.csv'):
    """Sweeps GoPro Photo Lineup.csv and returns dict of photo names + approximate locations of them in a given video
    
    Optional Inputs:
        path: local path that leads to an appropriately-formatted csv file
            
    Outputs:
        videomapping: A dict containing keys of every GoPro image in the Madagascar2018 folders, whose values are lists
            containing, in order: the gopro video where the gopro frame may be found, the approximate frame location of
            the photo, and the folder in dropbox corresponding to the date the video was taken (e.g. 10-17, 10-20 etc.)

            Video name and video date are output as strings, approximate frame location is an int

    Videomapping is meant to feed into findframe function below
    """
    
    csv_file = open(path)
    csv_reader = csv.reader(csv_file, delimiter =',')
    videomapping = {}
    for row in csv_reader:
        if row[0] == 'GoProVideoFiles': # Skip first row of csv bc it only contains column labels
            continue
        elif row[0]:
            videoname = row[0] # Grab video name
        timesplit = row[2].split(':') # Split up 
        frame_approx = (int(timesplit[0])*3600+int(timesplit[1])*60+int(timesplit[2]))*24
        videomapping[row[1]] = [videoname, frame_approx]#, folder]

    return(videomapping)
```

```python
### BEFORE RUNNING THIS CELL, POPULATE image_paths, video_paths, csv_path AT THE BOTTOM WITH YOUR FILEPATHS

def findframe(images, imagepath, videopath, csvpath, framecheck = 40, showgraph = True, resolution = (1920,1440)):
    """Function to find exact matching frame of a given image or list of images
    
    INPUTS:
        imagepaths: Path to image files being compared against videos
        videopath: Path to video files beign compared against images
        csvpath: path pointing to csv lineup document, passed into csvsweep()
        framecheck (optional, default 40): Number of frames to check around approximate image-frame location, as calculated by csvsweep
        showgraph (optional, default True): Boolean value to toggle graphins frame-matching with matplotlib
        resolution (optional, default (1920, 1440)): Resolution photo will be resized to, should be equal to video resolution
    
    OUTPUTS:
        bestmatches: Array of arrays containing GoPro images and their keypoints, matched frames and their keypoints,
            a set of best matched points between the image and frame, the frame number, the image file name, and the video file name
    
    TO USE:
        -Import all packages according to the top cell in this notebook
        -Run csvsweep cell above
        -Populate image_paths, video_paths, and csv_path appropriately at the bottom of this cell
        -Run this cell
    """
    
    # Run csvsweep to create dict of all photos and their approximate frame locations in corresponding videos
    videokey = csvsweep(csvpath)
    
    bestmatches = []
    #images = glob.glob(imagepath)
    index = 0
    
    for img in images:
        # Raise error if image input not in dict, otherwise establish videopath, imagepath, frameguess
        #imagefile = img.split('/')[-1]
        #if imagefile in videokey:
        if img in videokey:
            #videofile = videokey[imagefile][0]
            #frameguess = videokey[imagefile][1]
            #videopath_new = videopath + videokey[imagefile][0]
            #imagepath = imagepath
            videofile = videokey[img][0]
            frameguess = videokey[img][1]
            videopath_new = videopath + videokey[img][0]
            imagepath_new = imagepath + img
            print(frameguess, videopath)
            index += 1
        else:
            raise NameError('Image ' + img  + ' Not Found. Please check imagename and paths against file structure')
            index += 1

        # Call video and image with openCV, resize and colorcorrect image
        print(imagepath_new)
        video = cv2.VideoCapture(videopath_new)
        image = cv2.resize(cv2.cvtColor(cv2.imread(imagepath_new), cv2.COLOR_BGR2RGB), resolution)

        # Create empty lists to store frames, keypoints, descriptors, matches, and means
        startframe = frameguess-framecheck
        frameset_colorcorrect = []
        kp = []
        des = []
        matches = []
        matches_means = []

        # Set start frame for video
        video.set(cv2.CAP_PROP_POS_FRAMES, startframe)
        
        # Capture frames in range of csvsweeps' guessed frame location
        for i in range(framecheck*2):
            rval, frame = cv2.VideoCapture.read(video)
            frameset_colorcorrect.append(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        video.release()

        # Initiate ORB detector for feature matching
        orb = cv2.ORB_create()
        
        # Get keypoints and descriptors of GoPro Image, then for all captured frames
        kp_img, des_img = orb.detectAndCompute(image, None)

        for i in range(len(frameset_colorcorrect)):
            kp_vid, des_vid = orb.detectAndCompute(frameset_colorcorrect[i], None)
            kp.append(kp_vid)
            des.append(des_vid)

        # Make BFMatcher object
        bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)

        # Match Descriptors
        for i in range(len(des)):
            matches.append(bf.match(des_img,des[i]))
            matches[-1] = sorted(matches[-1], key = lambda x:x.distance)
            matches_means.append(sum(m.distance for m in matches[-1])/len(matches[-1]))

        # Find frame with minimum difference from GoPro image and log that frame
        framestop = matches_means.index(min(matches_means))
        matchset = matches[framestop]
        exactframe = framestop+startframe
        
        # Append list to bestmatches containing image, keypoints, frame, matches, frame number, image and video filenames
        bestmatches.append([image, kp_img, frameset_colorcorrect[framestop], 
                            kp[framestop], matchset, exactframe, img, videofile])
        #                    kp[framestop], matchset, exactframe, imagefile, videofile])
        
    # Graph everything
    if showgraph is True:
        mpl.rcParams['figure.figsize'] = [20.0, 18.0] # Figuresize parameters can be adjusted depending on number of images being displayed

        for image_video_pair in range(len(bestmatches)):
            dataset = bestmatches[image_video_pair]
            matched_image = dataset[0]
            matched_kp_img = dataset[1]
            matched_frame = dataset[2]
            matched_kp_frame = dataset[3]
            matched_matches = dataset[4]
            framename = dataset[5]
            imagename = dataset[6]
            videoname = dataset[7]
            imagematches = cv2.drawMatches(matched_image, matched_kp_img, 
                                           matched_frame, matched_kp_frame, matched_matches[:20], None, flags=2)
            plt.subplot(len(bestmatches), 1, image_video_pair+1)
            plt.imshow(imagematches)
            plt.title(imagename + ' matched with frame ' + str(framename) + ' of video ' + videoname)

    return(bestmatches)

imgs = ['G0332826.JPG','G0332839.JPG', 'G0342946.JPG', 'G0343000.JPG']
image_paths = '/home/devynn/github/Github/omurasWhaleAnalysis/example_data/images/gopro_images/'
video_paths = '/home/devynn/github/Github/omurasWhaleAnalysis/example_data/videos/'
csv_path = '/home/devynn/github/Github/omurasWhaleAnalysis/example_data/csv/lineup_test.csv'

matcher = findframe(imgs, image_paths, video_paths, csv_path)
```
