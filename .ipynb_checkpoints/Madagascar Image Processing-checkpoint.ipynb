```python
#Downsize img2_rgb to the same size as frame_colorcorrect
img2_resize = cv2.resize(img2_rgb, (1920, 1440))

#Brute-force matcher tests from docs.opencv.org. Uses img2_resize and frame_colorcorrect from above

orb = cv2.ORB_create()

kp1, des1 = orb.detectAndCompute(img2_resize, None)
kp2, des2 = orb.detectAndCompute(frame_colorcorrect, None)

# create BF Matcher object
bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)

# Match Descriptors
matches = bf.match(des1,des2)

# Sort them in the order of their distance
matches = sorted(matches, key = lambda x:x.distance)

print(m.distance for m in matches)

# Draw first 20 matches
imagematches = cv2.drawMatches(img2_resize, kp1, frame_colorcorrect, kp2, matches[:20], None, flags=2)
# Tutorial code excludes None in pos 6 but appears to be necessary after following a stackoverflow thread on the topic

mpl.rcParams['figure.figsize'] = [16.0, 12.0]
plt.imshow(imagematches)
```

```python
# Playing around with brute-force outputs
# DMatch.distance - Distance between descriptors. The lower, the better it is. <- This is the important value
# DMatch.trainIdx - Index of the descriptor in train descriptors
# DMatch.queryIdx - Index of the descriptor in query descriptors
# DMatch.imgIdx - Index of the train image

# Grab paths to photo and video + resize photo
PATH3 = '/mnt/c/Users/Devynn/Dropbox (Point Road Solutions)/PRS_Engineering/Madagascar2018/10-20/Bullfrog_camera/G0171996.JPG'
img2_resize = cv2.resize(cv2.cvtColor(cv2.imread(PATH3), cv2.COLOR_BGR2RGB),(1920, 1440))

PATH4 = '/mnt/c/Users/Devynn/Dropbox (Point Road Solutions)/PRS_Engineering/Madagascar2018/10-20/Bullfrog_camera/GOPR1968.MP4'
vid2 = cv2.VideoCapture(PATH4)

# Establish starting frame, framerange, and reset count to 0
frameflag = 3357 + 26
framerange = 10
count = 0

# frameset is empty before video sweep
frameset = []

# Grab appropriate frames from video capture; release video when done
while(True):
    rval, frame = cv2.VideoCapture.read(vid2)
    if count >= frameflag - framerange//2:
        frameset.append(frame)
    if count == frameflag + framerange//2:
        frameset.append(frame)
        break
    count +=1
vid2.release()
```

```python
#Create repeatedly used variables out of the loop
orb = cv2.ORB_create()
# Keypoints and descriptors of static image
keypoint1, descriptor1 = orb.detectAndCompute(img2_resize, 0)
# BF matcher object
bfmatcher = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)

#Check all frames for similarities to chosen image, then display the one with the lowest mean distance
 
keypoints = []
keypoint = 0
descriptors = []
descriptor = 0
setofmatches = []
means = []

# Sweep through frames and determine likelihood of frame match with mean of bfmatcher distance outputs
for i in range(len(frameset)-1):

    keypoint, descriptor = orb.detectAndCompute(cv2.cvtColor(frameset[i], cv2.COLOR_BGR2RGB), None)
    
    keypoints.append(keypoint)
    descriptors.append(descriptor)


    matches = bfmatcher.match(descriptor1,descriptor)

    # Sort them in the order of their distance
    matches = sorted(matches, key = lambda x:x.distance)
    # Evaluate 20 top matches
    setofmatches.append([m.distance for m in matches[:20]])
    means.append(sum(setofmatches[-1])/len(setofmatches[-1]))

print(min(means))    

# Get frame with min average distance value
minpos = means.index(min(means))
print(frameflag - framerange//2 + minpos)
#Does it work (Draws matches)
doesthiswork = cv2.drawMatches(img2_resize, keypoint1, cv2.cvtColor(frameset[minpos], cv2.COLOR_BGR2RGB), keypoints[minpos], matches[:20], None, flags=2)

# Plot matches with and without drawn lines
mpl.rcParams['figure.figsize'] = [16.0, 12.0]
#plt.subplot(1,2,1)
#plt.imshow(cv2.cvtColor(img2_resize, cv2.COLOR_BGR2RGB))
#plt.title('GoPro Photo')
#plt.subplot(1,2,2)
#plt.imshow(cv2.cvtColor(frameset[minpos], cv2.COLOR_BGR2RGB))
#plt.show('Best Match Frame')
plt.imshow(doesthiswork)

#lt.show()
```
